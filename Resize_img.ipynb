```python
# --- Install required library ---
!pip install pillow

import os, zipfile, random, string
from PIL import Image
from google.colab import files

# --- Step 1: Upload ZIP ---
print("üìÅ Please upload your ZIP file containing images:")
uploaded = files.upload()
zip_name = next(iter(uploaded))

# --- Step 2: Extract ZIP ---
input_dir = "/content/input_images"
os.makedirs(input_dir, exist_ok=True)
with zipfile.ZipFile(zip_name, "r") as zip_ref:
    zip_ref.extractall(input_dir)

# --- Step 3: Prepare output folder ---
output_dir = "/content/output_images"
os.makedirs(output_dir, exist_ok=True)

# --- Helper: Generate random names ---
def random_name(length=6):
    return ''.join(random.choices(string.ascii_lowercase, k=length))

# --- Step 4: Resize + convert to PNG with random names ---
size = (256, 256)
count = 0

for root, _, files_in_dir in os.walk(input_dir):
    for file in files_in_dir:
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff', '.webp')):
            try:
                img_path = os.path.join(root, file)
                img = Image.open(img_path).convert("RGB")
                img = img.resize(size, Image.LANCZOS)

                # Generate random name
                random_filename = f"{random_name()}.png"
                out_path = os.path.join(output_dir, random_filename)
                img.save(out_path, "PNG")
                count += 1
            except Exception as e:
                print(f"‚ö†Ô∏è Skipped {file}: {e}")

print(f"\n‚úÖ Processed {count} images ‚Üí saved as random PNGs in {output_dir}")

# --- Step 5: Create ZIP for download ---
output_zip = "/content/resized_random_images.zip"
with zipfile.ZipFile(output_zip, "w", zipfile.ZIP_DEFLATED) as zipf:
    for file in os.listdir(output_dir):
        zipf.write(os.path.join(output_dir, file), arcname=file)

print(f"üì¶ Created ZIP: {output_zip}")

# --- Step 6: Download result ---
files.download(output_zip)
```
